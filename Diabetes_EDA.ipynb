#2. Know About Your Data

1. **Pregnancies**: Number of times pregnant
2.**Glucose**: Plasma glucose concentration a 2 hours in an oral glucose
tolerance test
3. **BloodPressure**: Diastolic blood pressure (mm Hg)
4. **SkinThickness**: Triceps skin fold thickness (mm)
5. **Insulin**: 2-Hour serum insulin (mu U/ml)
6. **BMI**: Body mass index (weight in kg/(height in m)^2)
7.**DiabetesPedigreeFunction**: Diabetes pedigree function (a function which scores likelihood of diabetes based on family history)
8. **Age**: Age in years
9. **Outcome**: Class variable (0 or 1) indicating whether the person has diabetes or not

#5. Univarient Analysis

#Analysis Of Pregnancies

```python
#Sub plot of Pregnancies
fig, ax = plt.subplots(1,2,figsize = (20,8))
fig, ax1 = plt.subplots(figsize = (20,8))
sns.histplot(x = 'Pregnancies',kde = True,ax = ax[0],color = '#c107ff',data = df,edgecolor = 'Black') #Histplot of Pregnancies
sns.boxplot(x = 'Pregnancies',ax = ax[1],data = df,color = '#4507ff') #Boxplot of Pregnancies
sns.violinplot(x = 'Pregnancies',ax = ax1,data = df,color = '#c1ff07') #Violinplot of Pregnancies
```

```python
print("Median of Pregnancies: ",df['Pregnancies'].median())
print("Maximum of Pregnancies: ",df['Pregnancies'].max())
```

```python
df['Pregnancies'].value_counts()
```

From The Above Pregnancies Analysis, We Observe That

* The median of the Pregnancies is 3.0,
* The Maximum Pregnancies Happened are 17,
* Most Of The Patients Had either 0,1 or 2 Pregnancies.




#Analysis Of Outcome

```python
#Count plot of Outcome and Pie Chart
fig, ax = plt.subplots(1,2, figsize = (10,8))

sns.countplot(x = 'Outcome',data = df,color = '#4507ff',ax = ax[0],edgecolor = 'black')

df['Outcome'].value_counts().plot(kind = 'pie',explode = [0,0], wedgeprops ={'linewidth':1,'edgecolor':'black'},
                                 shadow = True, autopct = '%1.2f%%',ax = ax[1],labels = ['No','Yes'],colors = ['#ff4507','#ff0745'])
```

From The Above Outcome Analysis, We Observed That

* Around 65.1% Patients In The Dataset Do Not Have Diabetes.
* Around 34.90% Patients In The Dataset Have Diabetes.
